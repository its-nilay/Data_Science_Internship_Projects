<h1>Supermart Grocery Sales</h1>

```python
#Import libraries for data handling and visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

<h3>Load the CSV file and clean it up for analysis.</h3>

```python
#Load the dataset from the CSV file
data = pd.read_csv('Supermart Grocery Sales - Retail Analytics Dataset.csv')
# Display the first 5 rows to see the data 
data.head()
```

<h3>Check for missing values in each column</h3>

```python
#Check for missing values in each column
data.isnull().sum()
```

```python
#Drop any rows that have missing data
data.dropna(inplace=True)
```

```python
#Drop any duplicate rows
data.drop_duplicates(inplace=True)
```

```python
data.head()
```

```python
#Convert 'Order Date' column to a proper date format
# Convert 'Order Date', handling mixed formats like '11-08-2017' and '4/15/2018'
data['Order Date'] = pd.to_datetime(data['Order Date'], format='mixed')

# Create new columns for year and month name from 'Order Date'
data['year'] = data['Order Date'].dt.year
data['Month'] = data['Order Date'].dt.strftime('%B')
```

```python
data.head()
```

<h3>Creating visualizations to understand the sales data (EDA)</h3>

```python
# Set the figure size for better readability
plt.figure(figsize=(10, 6))

# Group data by 'Category' and sum the 'Sales' for each
# .sort_values(ascending=False) is added to make the chart easier to read
sales_by_category = data.groupby('Category')['Sales'].sum().sort_values(ascending=False).reset_index()

# Create the bar plot using seaborn
sns.barplot(x='Category', y='Sales', data=sales_by_category)

# Add title and labels for clarity
plt.title('Total Sales by Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')

# Rotate the x-axis labels to prevent them from overlapping
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(8, 6))

# Group data by 'year' and sum the 'Sales'
yearly_sales = data.groupby('year')['Sales'].sum().reset_index()

# Create the bar plot using seaborn
sns.barplot(x='year', y='Sales', data=yearly_sales)

# Add title and labels
plt.title('Total Sales by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')

# Display the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(10, 6))

# Group by city, sum sales, sort, and get the top 5
top_cities = data.groupby('City')['Sales'].sum().sort_values(ascending=False).head(5).reset_index()

# Create the bar plot using seaborn
sns.barplot(x='City', y='Sales', data=top_cities)

# Add title and labels
plt.title('Top 5 Cities by Sales')
plt.xlabel('City')
plt.ylabel('Total Sales')

# Rotate the x-axis labels to prevent overlap
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(10, 8))

# Calculate the correlation matrix for numeric columns only
corr_matrix = data.corr(numeric_only=True)

# Create the heatmap using seaborn
# annot=True displays the numbers on the map
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

# Add a title
plt.title('Correlation Heatmap')

# Display the plot
plt.show()
```
